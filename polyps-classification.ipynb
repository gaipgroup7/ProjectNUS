# 1. Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import cv2
import time
import seaborn as sns
from keras.preprocessing.image import ImageDataGenerator
from tqdm import tqdm
```

# 2. Data Preprocessing

```python
train_list_polyp = os.listdir('../input/PolypClassificationCleaned/Polyp Classification Cleaned/train/Polyp')
train_list_nopolyp = os.listdir('../input/PolypClassificationCleaned/Polyp Classification Cleaned/train/NonPolyp')
test_list_polyp = os.listdir('../input/PolypClassificationCleaned/Polyp Classification Cleaned/test/Polyp')
test_list_nopolyp = os.listdir('../input/PolypClassificationCleaned/Polyp Classification Cleaned/test/NonPolyp')
valid_list_polyp = os.listdir('../input/PolypClassificationCleaned/Polyp Classification Cleaned/valid/Polyp')
valid_list_nopolyp = os.listdir('../input/PolypClassificationCleaned/Polyp Classification Cleaned/valid/NonPolyp')
```

```python
x_train = np.zeros((len(train_list_polyp)*2, 128,128,3), dtype=np.uint8)
x_val = np.zeros((len(valid_list_polyp)*2, 128,128,3), dtype=np.uint8)
x_test = np.zeros((len(test_list_polyp)*2, 128,128,3), dtype=np.uint8)
y_train = np.zeros((len(train_list_polyp)*2), dtype=np.uint8)
y_val = np.zeros((len(valid_list_polyp)*2), dtype=np.uint8)
y_test = np.zeros((len(test_list_polyp)*2), dtype=np.uint8)
```

```python
x_train.shape
```

```python
int(len(train_list_nopolyp))
```

```python
for i in tqdm(range(len(train_list_nopolyp))):
    img =  tf.keras.preprocessing.image.load_img('../input/PolypClassificationCleaned/Polyp Classification Cleaned/train/NonPolyp/'+str(train_list_nopolyp[i]), target_size=(128,128), color_mode='rgb')
    x_train[i]=img
for i in tqdm(range(len(train_list_polyp),(2*len(train_list_polyp)))):
    img =  tf.keras.preprocessing.image.load_img('../input/PolypClassificationCleaned/Polyp Classification Cleaned/train/Polyp/'+str(train_list_polyp[i-int(len(train_list_nopolyp))]), target_size=(128,128), color_mode='rgb')
    x_train[i]=img
    y_train[i]=1
```

```python
for i in tqdm(range(len(valid_list_nopolyp))):
    img =  tf.keras.preprocessing.image.load_img('../input/PolypClassificationCleaned/Polyp Classification Cleaned/valid/NonPolyp/'+str(valid_list_nopolyp[i]), target_size=(128,128), color_mode='rgb')
    x_val[i]=img
for i in tqdm(range(len(valid_list_polyp),(2*len(valid_list_polyp)))):
    img =  tf.keras.preprocessing.image.load_img('../input/PolypClassificationCleaned/Polyp Classification Cleaned/valid/Polyp/'+str(valid_list_polyp[i-int(len(valid_list_nopolyp))]), target_size=(128,128), color_mode='rgb')
    x_val[i]=img
    y_val[i]=1
```

```python
for i in tqdm(range(len(test_list_nopolyp))):
    img =  tf.keras.preprocessing.image.load_img('../input/PolypClassificationCleaned/Polyp Classification Cleaned/test/NonPolyp/'+str(test_list_nopolyp[i]), target_size=(128,128), color_mode='rgb')
    x_test[i]=img
for i in tqdm(range(len(test_list_polyp),(2*len(test_list_polyp)))):
    img =  tf.keras.preprocessing.image.load_img('../input/PolypClassificationCleaned/Polyp Classification Cleaned/test/Polyp/'+str(test_list_polyp[i-int(len(test_list_nopolyp))]), target_size=(128,128), color_mode='rgb')
    x_test[i]=img
    y_test[i]=1
```

```python
x_train = x_train/255.0
x_val = x_val/255.0
x_test = x_test/255.0
```

```python
x_val.shape
```

# 3. Creating Models

# Model 1

```python
cnn1 = tf.keras.models.Sequential()
cnn1.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, activation = 'relu', input_shape = [128,128,3]))
cnn1.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
cnn1.add(tf.keras.layers.Dropout(0.20))
cnn1.add(tf.keras.layers.Conv2D(filters = 64, kernel_size = 3, activation = 'relu'))
cnn1.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
cnn1.add(tf.keras.layers.Dropout(0.20))
cnn1.add(tf.keras.layers.Conv2D(filters = 128, kernel_size = 3, activation = 'relu'))
cnn1.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
cnn1.add(tf.keras.layers.Dropout(0.20))
cnn1.add(tf.keras.layers.Conv2D(filters = 256, kernel_size = 3, activation = 'relu'))
cnn1.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
cnn1.add(tf.keras.layers.Dropout(0.20))
cnn1.add(tf.keras.layers.Flatten())
cnn1.add(tf.keras.layers.Dense(units = 64, activation = 'relu'))
cnn1.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
cnn1.compile(optimizer = 'adam', loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
cnn1.summary()
```

```python
history1 = cnn1.fit(x = x_train,y = y_train, validation_data = (x_val,y_val), epochs = 25, batch_size = 16)
```

# Model 4

```python
cnn4 = tf.keras.models.Sequential()
cnn4.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, activation = 'relu', input_shape = [128,128,3]))
cnn4.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
cnn4.add(tf.keras.layers.Dropout(0.20))
cnn4.add(tf.keras.layers.Conv2D(filters = 64, kernel_size = 3, activation = 'relu'))
cnn4.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
cnn4.add(tf.keras.layers.Dropout(0.20))
cnn4.add(tf.keras.layers.Conv2D(filters = 128, kernel_size = 3, activation = 'relu'))
cnn4.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
cnn4.add(tf.keras.layers.Dropout(0.20))
cnn4.add(tf.keras.layers.Conv2D(filters = 256, kernel_size = 3, activation = 'relu'))
cnn4.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
cnn4.add(tf.keras.layers.Dropout(0.20))
cnn4.add(tf.keras.layers.Flatten())
cnn4.add(tf.keras.layers.Dense(units = 64, activation = 'relu',kernel_regularizer=tf.keras.regularizers.L2(0.01)))
cnn4.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid',kernel_regularizer=tf.keras.regularizers.L2(0.01)))
cnn4.compile(optimizer = 'adam', loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
cnn4.summary()
```

```python
history4 = cnn4.fit(x = x_train,y = y_train, validation_data = (x_val,y_val), epochs = 25, batch_size = 16)
```
